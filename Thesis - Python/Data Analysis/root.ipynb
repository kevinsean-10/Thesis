```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import re
pd.options.display.min_rows = 10
plt.rcParams['font.family'] = 'Times New Roman'
plt.rcParams['font.size'] = 12  # You can also set the default font size
```

# Data 1

```python
data1 = pd.read_excel('Analisis Problem.xlsx',sheet_name="Problem 1 - Roots",header=1)
data1
```

```python
data1_rade = data1.iloc[:,:5]
data1_rade.rename(inplace=True,columns={'max_gen':"Maks Iter"})
data1_rade['percentage'] *= 100
popsize = [50,100,250]
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
cmap = sns.diverging_palette(145, 300, s=60,l=60,center='dark', as_cmap=True)
for id_popsize in range(len(popsize)):
    data1_rade_param = data1_rade.loc[(data1_rade['pop_size']==popsize[id_popsize])]
    sns.barplot(data1_rade_param,ax=axes[id_popsize],x="roots found", y="percentage",hue="Maks Iter",palette=cmap)
    axes[id_popsize].set(xlabel='Solusi yang ditemukan', ylabel='Persentase', title=f"Populasi {popsize[id_popsize]}")
    axes[id_popsize].set_ylim(0, 100)
fig.savefig(f'D:/OneDrive - Institut Teknologi Bandung/[AKADEMIK]/Semester 7-8/TA/Draft/Draft Buku/Gambar/Stat/Masalah 1/RADE, Masalah 1, Distribusi Akar.png')
```

```python
data1_hde = data1.iloc[:,6:11]
data1_hde.rename(inplace=True,columns={'max_gen.1':"Maks Iter"})
data1_hde['percentage.1'] *= 100
popsize = [50,100,250]
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
cmap = sns.diverging_palette(145, 300, s=60,l=60,center='dark', as_cmap=True)
for id_popsize in range(len(popsize)):
    data1_hde_param = data1_hde.loc[(data1_hde['parts']==popsize[id_popsize])]
    sns.barplot(data1_hde_param,ax=axes[id_popsize],x="roots found.1", y="percentage.1",hue="Maks Iter",palette=cmap)
    axes[id_popsize].set(xlabel='Solusi yang ditemukan', ylabel='Persentase', title=f"Populasi {popsize[id_popsize]}")
    axes[id_popsize].set_ylim(0, 100)
fig.savefig(f'D:/OneDrive - Institut Teknologi Bandung/[AKADEMIK]/Semester 7-8/TA/Draft/Draft Buku/Gambar/Stat/Masalah 1/HDE, Masalah 1, Distribusi Akar.png')
```

```python
data1_sdde = data1.iloc[:,12:]
data1_sdde.rename(inplace=True,columns={'k_max':"Maks Iter"})
data1_sdde['percentage.2'] *= 100
popsize = [50,100,250]
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
cmap = sns.diverging_palette(145, 300, s=60,l=60,center='dark', as_cmap=True)
for id_popsize in range(len(popsize)):
    data1_sdde_param = data1_sdde.loc[(data1_sdde['m_cluster']==popsize[id_popsize])]
    sns.barplot(data1_sdde_param,ax=axes[id_popsize],x="roots found.2", y="percentage.2",hue="Maks Iter",palette=cmap)
    axes[id_popsize].set(xlabel='Solusi yang ditemukan', ylabel='Persentase', title=f"Populasi {popsize[id_popsize]}")
    axes[id_popsize].set_ylim(0, 100)
fig.savefig(f'D:/OneDrive - Institut Teknologi Bandung/[AKADEMIK]/Semester 7-8/TA/Draft/Draft Buku/Gambar/Stat/Masalah 1/SDDE, Masalah 1, Distribusi Akar.png')
```
