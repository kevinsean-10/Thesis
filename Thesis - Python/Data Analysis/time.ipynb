```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
pd.options.display.min_rows = 10
plt.rcParams['font.family'] = 'Times New Roman'
plt.rcParams['font.size'] = 12  # You can also set the default font size
```

```python
data1 = pd.read_excel('Analisis Problem.xlsx',sheet_name="Problem 1 - Stat",header=1)
data1_rade = data1.iloc[:36]
data1_rade.rename(inplace=True, columns={'pop_size': "Population", "max_gen": "Maks Iter","time_elapsed":"Waktu" })
data1_hde = data1.iloc[39:74,:6]
data1_hde.columns = data1.iloc[38,:6]
data1_hde.reset_index(inplace=True, drop=True)
data1_hde.rename(inplace=True, columns={'pop_size': "Population", "max_gen": "Maks Iter","time_elapsed":"Waktu" })
data1_sdde = data1.iloc[78:,:6]
data1_sdde.columns = data1.iloc[77,:6]
data1_sdde.reset_index(inplace=True, drop=True)
data1_sdde.rename(inplace=True, columns={'pop_size': "Population", "max_gen": "Maks Iter","time_elapsed":"Waktu" })
```

```python
method1 = [data1_rade, data1_hde, data1_sdde]
titles = ['Metode RADE', 'Metode HDE', 'Metode Clustering DE']
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

cmap = sns.diverging_palette(145, 300, s=60,l=60,center='dark', as_cmap=True)
legend_titles = ["NP", "N", "m_cluster"]

for id_method in range(len(method1)):
    data1_method = method1[id_method][method1[id_method]['Stat'] == 'mean']
    sns.lineplot(data=data1_method, ax=axes[id_method], x='Maks Iter', y='Waktu', hue='Population', palette=cmap)
    # Modify the legend title
    handles, labels = axes[id_method].get_legend_handles_labels()
    axes[id_method].legend(handles=handles, labels=labels, title=legend_titles[id_method])
    axes[id_method].set_title(titles[id_method])

# fig.suptitle('Komparasi Waktu terhadap Banyak Generasi Masalah 1', fontsize=16)
plt.tight_layout()
fig.savefig(f'D:/OneDrive - Institut Teknologi Bandung/[AKADEMIK]/Semester 7-8/TA/Draft/Draft Buku/Gambar/Stat/Masalah 1/Perbandingan Waktu Masalah 1.png')
plt.show()
```
