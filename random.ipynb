```python
import matplotlib.pyplot as plt
import numpy as np

# Define cluster centers and radii
cluster_centers = [(2, 2), (5, 5), (8, 2)]
cluster_radii = [1, 1.5, 0.8]

# Create a figure and axis
fig, ax = plt.subplots()

# Plot the clusters as circles
for center, radius in zip(cluster_centers, cluster_radii):
    circle = plt.Circle(center, radius, fill=False, linestyle='dotted', edgecolor='b')
    ax.add_artist(circle)

# Set axis limits
ax.set_xlim(0, 10)
ax.set_ylim(0, 10)

# Add labels (optional)
for i, center in enumerate(cluster_centers):
    ax.text(center[0], center[1], f'Cluster {i+1}', ha='center', va='bottom')

# Add a title and labels (optional)
ax.set_title('Cluster Visualization')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')

# Show the plot
plt.gca().set_aspect('equal', adjustable='box')  # Make the aspect ratio equal
plt.grid(True)
plt.show()
```
